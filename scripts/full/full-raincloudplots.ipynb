```python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ptitprince as pt
import researchpy as rp
import scipy.stats as stats

sns.set_context("talk")
```

### Get Metrics

```python
masterData = pd.read_csv('../../exports/masterData.csv')
```

```python
masterData['mratio'] = masterData['metad']/masterData['dprime']
masterData['mdiff'] = masterData['metad']-masterData['dprime']
masterData['mratio_log'] = np.log(masterData['mratio'])
```

```python
# get rid of studies with atypical subjects
masterData = masterData[masterData.domain != 'Asd'] # wang
masterData = masterData[masterData.domain != 'Patient'] # sadeghi
masterData = masterData[masterData.domain != 'Visuomotor'] # arbuzova
masterData = masterData[masterData.domain != 'Motor'] # arbuzova
masterData = masterData[masterData.domain != 'Motor'] # arbuzova

# drop the max outlier
masterData = masterData[masterData.mratio_log != 16.412810209706187]
```

```python
# collapse to perception/memory
masterData.loc[masterData.dataset == 'Matthews_2019', 'domain'] = "Memory"
masterData.loc[masterData.dataset == 'Sadeghi_2017_memory', 'domain'] = "Memory"
masterData.loc[masterData.dataset == 'Clark_2018', 'domain'] = "Perception"
masterData.loc[masterData.dataset == 'Bang_2019_Exp2', 'domain'] = "Perception"
masterData.loc[masterData.dataset == 'Haddara_2022_Expt1', 'domain'] = "Perception"
masterData.loc[masterData.dataset == 'Song_2011', 'domain'] = "Perception"
masterData.loc[masterData.dataset == 'Wang_2017_Neuropsychologia', 'domain'] = "Perception"
masterData.loc[masterData.dataset == 'Sadeghi_2017_perception', 'domain'] = "Perception"
masterData.loc[masterData.domain == 'Wm', 'domain'] = "Memory"
masterData.loc[masterData.domain == 'Oriwm', 'domain'] = "Memory"
masterData.loc[masterData.domain == 'Percept', 'domain'] = "Perception"
masterData.loc[masterData.domain == 'Visual', 'domain'] = "Perception"
masterData.loc[masterData.domain == 'Oriperc', 'domain'] = "Perception"
masterData.loc[masterData.domain == 'Contrast', 'domain'] = "Perception"
```

# Stats

### log M-ratio

```python
summary, results = rp.ttest(group1= masterData['mratio_log'][masterData['domain'] == 'Memory'], group1_name= "Memory",
         group2= masterData['mratio_log'][masterData['domain'] == 'Perception'], group2_name= "Perception")

print(summary)
print(results)

statistic, pvalue = stats.ttest_ind(masterData.dropna()['mratio_log'][masterData.dropna()['domain'] == 'Memory'],
         masterData.dropna()['mratio_log'][masterData.dropna()['domain'] == 'Perception'])

print('T =', statistic)
print('p =', pvalue)
```

### M-diff

```python
summary, results = rp.ttest(group1= masterData['mdiff'][masterData['domain'] == 'Memory'], group1_name= "Memory",
         group2= masterData['mdiff'][masterData['domain'] == 'Perception'], group2_name= "Perception")

print(summary)
print(results)

statistic, pvalue = stats.ttest_ind(masterData['mdiff'][masterData['domain'] == 'Memory'],
         masterData['mdiff'][masterData['domain'] == 'Perception'])

print('T =', statistic)
print('p =', pvalue)
```

### Raincloud Plots (All)

```python
sns.set(style="whitegrid", font_scale=2.5)
def plotFullRaincloud(data, x):
    """
    Create raincloud plots with input = {data} and {x}
    """
    
    colors = ['#0ABAF9' if ('Memory' in y) else '#E5313D' for y in data['domain'].unique()]
    
    f, ax = plt.subplots(figsize=(30, 20))
    
    ax = pt.half_violinplot(x=x, y='domain', data=data, palette=colors,
                           bw=0.2, cut=0, scale='area', width=.6, inner=None, orient='h',linewidth=4)
    # ax.axvline(0, ls='--', color='orange')
    ax = sns.stripplot(x=x, y='domain', data=data, palette=colors, edgecolor='white', size=4, jitter=1,
                      zorder=0, orient='h')
    ax = sns.boxplot(x=x, y='domain', data=data, color='black', width=.15, zorder=10, showcaps=True,
                     boxprops={'facecolor': 'none', 'zorder': 10}, showfliers=True,
                     whiskerprops={'linewidth': 2, 'zorder': 10}, saturation=1, orient='h')

    plt.savefig('figure1_mratio_log.pdf', format='pdf', dpi=1200)
```

```python
plotFullRaincloud(data=masterData, x='mratio_log')
```

```python
plotFullRaincloud(data=masterData, x='mdiff')
```
